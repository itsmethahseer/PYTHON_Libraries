```python
# but we can't replace by zero with all the rows.So we use python dictionaries for filling diff values with diff columns.
new_df=df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no event'
})
new_df
```

```python
# we can fill empty values from the next values by method=bfill
new_df=df.fillna(method="bfill")
new_df
```

```python
#if you want to fill along columns
new_df=df.fillna(method='bfill',axis="columns")
new_df
```

```python
# if you want to fill only one previous value,you can use limit.
new_df=df.fillna(method='ffill',limit=1)
new_df
```

# Merge two dataframes

```python
df1=pd.DataFrame({
    'city':['mumbai','delhi','banglure'],
    'temperature':[23,45,34]
})
df2=pd.DataFrame({
    'city':['mumbai','delhi','banglure'],
    'temperature':[34,37,39]
})
```

```python
df1
```

```python
df2
```

```python
df=pd.merge(df1,df2,on="city")
df
```

```python
#default the merge will take only common values(like intersection in sets)
df1=pd.DataFrame({
    'city':['new york','chichgo','orlando','baltimore'],
    'temperature':[21,23,14,19]
})
df2=pd.DataFrame({
    'city':['chichgo','new york','San fransisco'],
    'humidiy':[45,56,45]
})
```

```python
df1
```

```python
df2
```

```python
df=pd.merge(df1,df2,on="city")
df
```

```python
# to avoid this
df=pd.merge(df1,df2,on="city",how="outer")
df
# default the value of the how is inner(like intersection in sets)
```

```python
df=pd.merge(df1,df2,on="city",how="left")
df
# it will print the first dataframe with common elements
```

```python
df=pd.merge(df1,df2,on="city",how="right")
df
# it will print the second dataframe with common elements
```

```python
df=pd.merge(df1,df2,on="city",how="outer",indicator=True)
df
# the indicator will give you the dataframes identity, data from left,right,both etc.
```
