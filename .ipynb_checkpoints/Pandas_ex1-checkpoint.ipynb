```python
# for printing the n.a. values form the column people as sam walton.
import pandas as pd
def convert_people_cell(cell):
    if cell=='n.a.':
        return 'sam walton'
    return cell
def convert_eps_cell(cell):
    if cell=='not avalilable':
        return 'None'
    return cell  
df=pd.read_excel('C:/Users/user/Downloads/mynew.xlsx',converters={
    'people':convert_people_cell,
    'eps':convert_eps_cell
})
df
```
