```python
import matplotlib.pyplot as plt
import numpy as np
xpoints=np.array([1,2,3,4])
ypoints=np.array([5,3,3,4])

plt.plot(xpoints,ypoints,'o')
```

```python
x=np.array([1,0,3,0,5])
plt.plot(x,marker="*")
```

```python
plt.plot(x,linestyle="-",c='red',lw=1)
```

```python
y=np.array([5,3,3,1,8])
plt.plot(x)
plt.plot(y)
```

```python
plt.plot(x)
plt.plot(y)
plt.xlabel("frequency")
plt.ylabel("velocity")
plt.title("Physics")
```

```python
plt.plot(x)
plt.plot(y)
plt.grid()
```

```python
plt.plot(x)
plt.plot(y)
plt.grid(axis="x")
```

```python
x=np.array([1,2,3,4,5])
y=np.array([4,1,6,3,2])
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("income")
x1=np.array([0,2,1,4,1])
y1=np.array([1,1,0,0,2])
plt.subplot(1,2,2)
plt.plot(x1,y1)
plt.title("sales")
plt.suptitle("My shop")
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.scatter(x1,y1)
```

```python
x=np.array([1,2,3,4,5])
y=np.array([4,1,0,2,8])
colors=np.array(["red",'violet','black','blue','yellow'])
plt.scatter(x,y,c=colors)
```
