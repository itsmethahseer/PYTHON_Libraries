```python
# Import seaborn
import seaborn as sns

# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

In the above code contain set_theme() used to set the default theme of seaborn.
using load_dataset() we create a dataframe namely tips. from the dataset 'tips'. it is a dataset in the seaborn for understanding the concepts.
-->relplot() is a function in the Seaborn library used to create a figure with one or more subplots.The subplots are created based on the columns passed as arguments to the function.
The hue argument is used to differentiate between smokers and non-smokers, while the style argument is used to show different marker styles for smokers and non-smokers. The size argument is used to scale the size of the markers based on the size of the party.


```python
import seaborn as sns
sns.set_theme()
tips = sns.load_dataset("tips")
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker")
```

function call produces two separate scatter plots, one for Lunch and one for Dinner. Within each plot, the points are colored based on whether the customer is a smoker or not, and a linear regression line is plotted through the points. This allows you to see if there is a correlation between the total_bill and tip variables for different times of day and for smokers versus non-smokers.

```python
import seaborn as sns
sns.set_theme()
fmri = sns.load_dataset("fmri")
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", col="region",
    hue="event", style="event",
)
```

```python
import seaborn as sns
sns.set_theme()
dots = sns.load_dataset("dots")
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate", col="align",
    hue="choice", size="coherence", style="choice",
    facet_kws=dict(sharex=False),
)
```

kind="line": This tells Seaborn to create a line plot.
facet_kws=dict(sharex=False): This specifies that each column of the plot should have its own x-axis scale, rather than sharing one x-axis across all columns.

```python
import seaborn as sns
sns.set_theme()
tips = sns.load_dataset("tips")
sns.displot(data=tips, x="total_bill", col="time", kde=True)
```

sns.displot(data=tips, x="total_bill", col="time", kde=True) creates a histogram of the total_bill variable in the tips dataset.
The kde parameter adds a smooth curve to the histogram that represents the estimated probability density function of the data.

```python
import seaborn as sns
sns.set_theme()
tips = sns.load_dataset("tips")
sns.catplot(data=tips, kind="swarm", x="day", y="total_bill", hue="smoker")
```

("tips" dataset, which contains information about the total bill, tip, sex, smoker status, day of the week, time of day, and size of the party for customers at a restaurant.) Here kind is 'swarm'.
catplot() is a function in the Python visualization library seaborn that is used to create categorical plots. It is a high-level interface for creating different types of categorical plots such as stripplot(), swarmplot(), boxplot(), violinplot(), boxenplot(), and pointplot().
## Plots for catogoracal data

```python
import seaborn as sns
sns.set_theme()
tips = sns.load_dataset("tips")
sns.catplot(data=tips, kind="violin", x="day", y="total_bill", hue="smoker",split=True)
```

The split parameter is set to True to split the violin plot by the "hue" category, i.e., "smoker" in this case. This creates two half-violins for each day, one for smokers and one for non-smokers.

```python
import seaborn as sns
sns.set_theme()
tips = sns.load_dataset("tips")
sns.catplot(data=tips, kind="bar", x="day", y="total_bill", hue="smoker")
```

```python
import seaborn as sns
sns.set_theme()
penguins = sns.load_dataset("penguins")
sns.jointplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species")
```

The sns.jointplot() function creates a joint plot, which combines a scatterplot and a histogram. It shows the relationship between two variables (x and y) and their distributions. The data argument specifies the dataset to use for plotting, while the x and y arguments specify the column names for the x and y variables. The hue argument is used to differentiate the data points based on a categorical variable (in this case, the "species" column).

So the complete code will create a joint plot of the "flipper_length_mm" and "bill_length_mm" variables from the "penguins" dataset, with each species represented by a different color.

```python
import seaborn as sns
sns.set_theme()
penguins = sns.load_dataset("penguins")
sns.pairplot(data=penguins, hue="species")
```

sns.pairplot(data=penguins, hue="species") is a function call that creates a pairplot of the penguin dataset. A pairplot is a type of plot that shows the pairwise relationships between variables in a dataset. In this case, the data parameter specifies the penguin dataset, and the hue parameter specifies the variable to use for coloring the points in the plot based on the different penguin species. The resulting plot will show scatterplots of each variable against every other variable in the dataset, with points colored based on their species.
