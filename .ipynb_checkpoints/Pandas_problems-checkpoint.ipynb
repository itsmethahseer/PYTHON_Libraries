```python
# if you want to write two dataframes into same excel file(different sheets)
df1={
    "name":['Thahseer','Lesin','Midhun','Nadeem'],
    'job':['data scientist','data analyst','data engineer','banker'],
    'Mob':[83748758,389595434,3498954984,3784733]
}
df2={
    "name":['robin','aswin','safeer','salman'],
    'job':['navy officer','socielist','data analyst','army'],
    'Mob':[83748758,389595434,3498954984,83784748]
}
```

```python
d1=pd.DataFrame(df1)
d2=pd.DataFrame(df2)
with pd.ExcelWriter("Details.xlsx") as writer:
    d1.to_excel(writer,sheet_name="stocks1")
    d2.to_excel(writer,sheet_name="stocks2")
```
